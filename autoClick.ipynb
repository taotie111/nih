```python
import pyautogui
import time
import keyboard
import sys

pyautogui.FAILSAFE = True

paused = False
running = True

def toggle_pause():
    global paused
    paused = not paused
    print("â¸ å·²æš‚åœ" if paused else "â–¶ ç»§ç»­æ‰§è¡Œ")

def exit_program():
    global running
    running = False
    print("ğŸ›‘ å·²æ”¶åˆ°é€€å‡ºæŒ‡ä»¤ï¼Œç¨‹åºç»“æŸ")

# ç»‘å®šçƒ­é”®ï¼ˆåªèƒ½ç»‘å®šä¸€æ¬¡ï¼‰
keyboard.add_hotkey('F8', toggle_pause)
keyboard.add_hotkey('esc', exit_program)

def wait_if_paused():
    """çœŸæ­£çš„æš‚åœæ§åˆ¶ç‚¹"""
    while paused:
        time.sleep(0.1)

# ======================
# ä¸»é€»è¾‘
# ======================

for i in range(40):
    if not running:
        break

    print(f"ç¬¬ {i+1} æ¬¡ç‚¹å‡»")

    wait_if_paused()
    pyautogui.click(1980, 1200)

    wait_if_paused()
    pyautogui.click(2040, 1200)

    time.sleep(0.1)

print("ç¨‹åºå·²å®‰å…¨é€€å‡º")
sys.exit(0)
```
